# set method

**set.add**

```python
s={1,2,3}
s.add(4)
s
```

**set.update()**

```python
s1={10,20,30,40,5}
s2={4,5}
s1.update(['a','c','f'])
print(s1)
s2
```

```python
a={1,2,3}
a.update([20,30,40])
a
```

**set.discard()**

```python
s={1,2,3}
s.discard(2)
s
```

```python
s={1,2,3}
s.discard(7)
s
```

**set.remove()**

```python
s={1,2,3}
s.remove(2)
s
```

```python
s={1,2,3}
s.remove(7)
s
```

**set.clear()**

```python
s={1,2,3}
s.clear()
s
```

**set.delete()**

```python
s={1,2,3}
del s
s
```

**set.copy()**

```python
s1={1,2,3}
s2=s1.copy()
s2.add(7)
print(s2)
s1
```

**set.union()**

```python
setA={1,2,3}
setB={2,3,4}
setA.union(setB)
```

**set.intersection()**

```python
setA={1,2,3}
setB={2,3,4}
setA.intersection(setB)
```

**set.isdisjoint()**

```python
setA={1,2,3}
setB={4,5,6}
setC={7,5,9}
s=setA.isdisjoint(setB)
print(s)

s=setB.isdisjoint(setC)
s
```

**set.symmetric_difference()**

```python
setA={1,2,3}
setB={2,3,4}
setA.symmetric_difference(setB)
```

**set.difference()**


```python
a={1,2,3}
b={2,3,4}
print(a-b)
print(b-a)
```

**issupset**

```python
a={3,4,5}
b={3,4,5,6}
ans=a.issubset(b)
ans
```

**issuperset**

```python
s1={3,4,5}
s2={3,4,5,6}
s=s1.issuperset(s2)
print(s)
s=s2.issuperset(s1)
s
```
