```python
# reverse the number
print("enter # when you enter all the number which you want to reverse")
l=[]
while True:
    n=input("enter the number:")
    if n=="#":
        break
    l.append(n)
l=l[::-1]
print(l)
```

```python
# reverse the digit
n=int(input("enter a number"))
rev=0
while n>0:
    digit=n%10
    rev=rev*10+digit
    n=n//10
print(rev)
```

```python
# fibonacci sequence
num1=0
num2=1
i=2
print(num1)
print(num2)
while i<=8:
    
    n=num1+num2
    print(n)
    num1=num2
    num2=n
    i+=1
    
```
