**convert a list of tuple into a dictionary**

```python
family=dict([('dad','business'),('mom','housewife'),('size',6)])
print(family)
```

```python
family=dict(('dad','business'),('mom','housewife'),('size',6))
print(family)
```

**INDEXING IN DICTIONARY**

##### SYNTEX-: dict[key name]

```python
cars={'car1':1989,'car2':[1,2,4],'car3':{110:3}}
print(cars['car1'])

print(cars['car2'][0])

print(cars['car3'][110])

```

## DICTIONARY METHOD

**dict.items()**

```python
marks={'history':45,'hindi':56}
marks.items()
```

**dict.keys()**

```python
marks={'history':45,'hindi':56}
marks.keys()
```

**dict.values()**

```python
marks={'history':45,'hindi':56}
marks.values()
```

**to add a single item into a dictionary**

```python
# option 1
marks={'history':45,'hindi':56}
marks['english']=90
marks
```

```python
# option 2
marks={'history':45,'hindi':56}
marks.update({'english':90})
print(marks)
```

**to add a single or multiple items into a dictionary**

```python
marks={'history':45,'hindi':56}
marks.update({'chemistry':93,"physic":78})
marks
```

**To delete item in dictionary**

```python
# option 1
marks={'history':45,'hindi':56}
del marks["history"]
marks
```

```python
marks={'history':45,'hindi':56}
del marks["his"]
marks
```

```python
# option 2
marks={"english":90,"hindi":93}
marks.pop("english")
marks
```

**dict.clear()**

```python
marks={'history':45,'hindi':56}
marks.clear()
marks
```

**dict.copy()**

```python
marks={'history':45,'hindi':56}
a=marks.copy()
a.update({'maths':94})
print(a)
marks
```

```python
marks={'history':45,'hindi':56}
marks.get("history")
```

```python
marks={'history':45,'hindi':56}
marks.get("his")
```

**replace a value in dictionary**

```python
marks={'history':45,'hindi':56}
marks["hindi"]=65
marks
```

**to replace the key**

```python
marks={'history':45,'hindi':56}
marks["maths"]=marks.pop("hindi")
marks
```
