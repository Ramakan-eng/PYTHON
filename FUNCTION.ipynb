```python
# upper lower 
def upper_lower(x):
    """
    Returns the upper and lower version of the string.
    The value must be a string, else it will result in
    an error.
    This function does not implement any error handling
    mechanism.
    """
    upper=x.upper() # convert x to upper string
    lower=x.lower() #convert x to lower string
    return upper ,lower # return both variable

upper,lower=upper_lower("python")
print(upper)
print(lower)
   
```

**Arbitary Arguments**


```python
def greet(*n):
    for i in n:
        print("hello",i)

# calling the function 
greet(1,2,3,4)
```

```python
def sum_all(*args):
    """ sum all value in the *args"""
    # initialize result 0
    result=0
    #sum all value
    for i in args:
        result+=i
    # print(result)
    return result

# calling the sum_all function with arbitrary number of argument
print(sum_all(1,2,3,4,5))
```

```python
# calling with different number of argument
print(sum_all(15,20,6))
```

```python
# ** use 

def info(**kwargs): #keyword argument
    """ print out key value pairs in **kwargs"""
    # Run for loop to prints dictionary items
    for key,value in kwargs.items():
        print(key+':'+ value)

# calling the function 

print(info(name="RK",age='20',residence="India"))

```

```python
print(info.__doc__) # docstring printing code
```

**Recussive Function or Recursion**

```python
def factorial(x):
    if type(x)==int:
        if x==0:
            return 1
        elif x>0:
            return x*factorial(x-1)
    else:
        print("value error")

# call the funtion 
n=int(input("enter the number"))

factorial(n)
        
    
```

**Anonymous(Lambda) Function**

```python
# define a function the "usual" way
def squared(x):
    return x**2

squared(2)
```

```python
# define an identical funtion using lambda
squared=lambda x:x**2

squared(2)
```
