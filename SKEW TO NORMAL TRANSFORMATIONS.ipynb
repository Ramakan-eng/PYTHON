```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Original right-skewed data
x = np.array([1, 2, 4, 7, 15, 30, 50, 100])

# Apply square root transformation
x_transformed = np.sqrt(x)

# Plot the original and transformed data
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Original Data
ax[0].hist(x, bins=8, color='blue', alpha=0.7)
ax[0].set_title("Original Right-Skewed Data")
ax[0].set_xlabel("Value")
ax[0].set_ylabel("Frequency")

# Transformed Data
ax[1].hist(x_transformed, bins=8, color='green', alpha=0.7)
ax[1].set_title("Transformed Data (Square Root)")
ax[1].set_xlabel("Value")
ax[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()

# Check normality using a statistical test
original_stat, original_p = stats.shapiro(x)
transformed_stat, transformed_p = stats.shapiro(x_transformed)

print("Original Data: Shapiro-Wilk Test p-value =", original_p)
print("Transformed Data: Shapiro-Wilk Test p-value =", transformed_p)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import boxcox, shapiro

# Original right-skewed data
x = np.array([1, 2, 4, 7, 15, 30, 50, 100])

# Apply Box-Cox transformation
x_transformed, lambda_opt = boxcox(x)

# Plot original and transformed data
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Original Data
ax[0].hist(x, bins=8, color='blue', alpha=0.7)
ax[0].set_title("Original Right-Skewed Data")
ax[0].set_xlabel("Value")
ax[0].set_ylabel("Frequency")

# Transformed Data
ax[1].hist(x_transformed, bins=8, color='green', alpha=0.7)
ax[1].set_title("Box-Cox Transformed Data")
ax[1].set_xlabel("Value")
ax[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()

# Check normality using Shapiro-Wilk test
original_stat, original_p = shapiro(x)
transformed_stat, transformed_p = shapiro(x_transformed)

print("Original Data: Shapiro-Wilk Test p-value =", original_p)
print("Transformed Data: Shapiro-Wilk Test p-value =", transformed_p)
print("Optimal  (lambda):", lambda_opt)
```

```python
import numpy as np
from scipy.stats import boxcox

# Example data (must be positive)
data = np.array([2, 3, 4, 5, 6])

# Perform the Box-Cox transformation and estimate lambda
transformed_data, lambda_val = boxcox(data, lmbda=None)

print("Transformed Data:", transformed_data)
print("Estimated Lambda:", lambda_val)
```
