**Random Module**

```python
import random 
```

```python
# random() function: this function always generate some float value between 0 and 1
random.random()
```

```python
# ranint() function to generate random integer between two given number(inclusive)
random.randint(1,100)
```

```python
# randrange() function: To generate random integer or integer in guven range
random.randrange(2,50,2)
```

```python
# choice() function :it won't return random number.It will return a random object from the given list
list=['sunny','bunny','chinny','vinny']

random.choice(list)
```
